## 0. Import Library

```python
import os
import math
import re
from collections import Counter
```

## 1. Load Document & Pre-processing
Load corpus dari folder data serta preprocessing dengan case-lowering, menghapus tanda baca dan men-tokenisasi.

Sumber: [Pandangan Hidup - Tan Malaka (1948)](https://www.marxists.org/indonesia/archive/malaka/1948-Pandangan.htm)

```python
def load_corpus_from_folder(folder_path):
    corpus = []
    for filename in sorted(os.listdir(folder_path)):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
                corpus.append(f.read().strip())
    return corpus

corpus = load_corpus_from_folder("data")


def tokenize(doc):
    doc = doc.lower()
    doc = re.sub(r'[^a-z\s]', '', doc)
    tokens = doc.split()
    return tokens

tokenized_corpus = [tokenize(doc) for doc in corpus]

for i, doc in enumerate(tokenized_corpus):
  print(f"Doc {i}: {doc}")
```

## 2. Compute Term Frequency

Menghitung seberapa sering sebuah kata muncul dalam satu dokumen.

```python
def compute_tf(tokenized_doc):
  total_terms = len(tokenized_doc)
  term_counts = Counter(tokenized_doc)
  tf = {}
  for term, count in term_counts.items():
    tf[term] = count / total_terms
  return tf

tf_all_docs = [compute_tf(doc) for doc in tokenized_corpus]

print("TF Result:")
for term, score in tf_all_docs[0].items():
  print(f" '{term}': {score:.4f}")
```

## 3. Compute Inverse Document Frequency

Mengukur seberapa jarang sebuah kata muncul di seluruh kumpulan dokumen.


```python
def compute_idf(tokenized_corpus):
  N = len(tokenized_corpus)

  document_frequency = {}
  for doc in tokenized_corpus:
    unique_terms = set(doc)
    for term in unique_terms:
      document_frequency[term] = document_frequency.get(term, 0) + 1

  idf = {}
  for term, freq in document_frequency.items():
    idf[term] = math.log(N / freq)

  return idf

idf_scores = compute_idf(tokenized_corpus)

print("IDF Result:")
for term, score in sorted(idf_scores.items(), key=lambda x: x[1]):
  print(f" '{term}': {score:.4f}")
```
